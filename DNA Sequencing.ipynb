```python
def AT_Counter (DNA, sig_figures):
    AT = (DNA.count("A") + DNA.count("T"))/len (DNA)
    return (round(AT, sig_figures))
print(AT_Counter("ATGC", 2))
print(AT_Counter(MySeq, 3))
print(AT_Counter("ATGCATGGTCCAGTCACTGCAAATC", 4))
```
